```python
# import libraries
import csv
import matplotlib.pyplot as plt 
```

```python
# Initiate empty lists
x = [] 
y = [] 
x1 = []
y1 = []
y2 = []
y3 = []
y4 = []
y5 = []

# import the data downloaded from Mongodb in CSV
with open('data.csv','r') as csvfile: 
    plots = csv.reader(csvfile, delimiter = ',') 
    next(plots) 
    # Loop through the collected data and populate the lists.
    for row in plots: 
        if int(row[5]) > 5000:
            x.append(row[2]) 
            y.append(int(row[4]))
            x1.append(row[3])
            y1.append(int(row[5]))
            y2.append(int(row[6]))
            y3.append(int(row[7]))
            y4.append(int(row[8]))
            y5.append(int(row[9]))
          
#Show the ages with the highest income.
# A column bar chart for Ages with Total Income greater than $5,000.
plt.figure(figsize=(6, 5))  
plt.bar(x, y, color = 'g', width = 0.72, label = "Age") 
plt.xlabel('Age') 
plt.ylabel('Income(USD)') 
plt.title('Ages with Income above $5,000') 
plt.legend() 
plt.show() 

```

### Explanation.
#### The visual above shows that people with the highest Income with age 55 being the highest.

```python
# Show the gender distribution across spending categories.
# Export the charts for use in a PowerPoint presentation for client use.
plt.figure(figsize=(14,8))
plt.subplot(2,3,1)
plt.bar(x1, y, color = 'b', width = 0.95, label = "Total Income") 
plt.title('Gender distribution on Total Income') 
plt.legend()
plt.savefig("GenderbyIncome.jpg")
plt.subplot(2,3,2)
plt.bar(x1, y1, color = 'r', width = 0.95, label = "Utilities") 
plt.title('Gender distribution on Utilities') 
plt.savefig("GenderbyUtilities.jpg")
plt.legend() 
plt.subplot(2,3,3)
plt.bar(x1, y2, color = 'g', width = 0.95, label = "Entertainment") 
plt.title('Gender distribution on Entertainment')
plt.savefig("GenderbyEntertainment.jpg")
plt.legend() 
plt.subplot(2,3,4)
plt.bar(x1, y3, color = 'y', width = 0.95, label = "School Fees") 
plt.title('Gender distribution on School Fees') 
plt.savefig("GenderbySchoolFees.jpg")
plt.legend() 
plt.subplot(2,3,5)
plt.bar(x1, y4, color = 'c', width = 0.95, label = "Shopping") 
plt.title('Gender distribution on Shopping') 
plt.savefig("GenderbyShopping.jpg")
plt.legend() 
plt.subplot(2,3,6)
plt.bar(x1, y5, color = 'hotpink', width = 0.95, label = "Healthcare") 
plt.title('Gender distribution on Healthcare')
plt.savefig("GenderbyHealthcare.jpg")
plt.legend() 
plt.show() 
```

### Explanation
#### The distribution visualizations above show that Males earn more and spend more than Females in all the spending categories.
